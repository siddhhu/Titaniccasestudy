```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
train=pd.read_csv("train.csv")    #import train dataset also already imported so nothing to do more
train.head()
```

```python
train.corr()    #not so co related
```

```python
train.info()
```

```python
train["Age"]=train["Age"].fillna(value=train["Age"].median())
```

```python
train['Embarked']=train['Embarked'].fillna(value='S')
train.isna().sum()
```

```python
train=train.drop(['Embarked'],axis=1)
print(train)
```

```python
# train
dummies_train=pd.get_dummies(train)
dummies_train
```

```python
dummies_train.columns
```

```python
features=list(set(dummies_train)-set(['Survived']))
features
```

```python
y=dummies_train['Survived'].values                 #storing output variabe in y
print(y)
```

```python
x=dummies_train[features].values                 #storing output variabe in x
print(x)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
logistic=LogisticRegression() 
```

```python
logistic.fit(train_x,train_y)
```

```python
prediction=logistic.predict(test_x)       #prediction from the test data 
prediction
```

```python

confusion_matrix=confusion_matrix(test_y, prediction)     #confusion matrix   
confusion_matrix
```

```python
accuracy_score=accuracy_score(test_y, prediction)        #calculating the accuracy
print(accuracy_score)                                    #This is the accuracy it can be mor accurate if we romove some more 
                             
```

```python
print("No of misclassified value :",(test_y!=prediction).sum())
```
